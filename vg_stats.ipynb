```python
import pandas as pd
g = pd.read_csv('vgsales.csv')
```

#### Most common publisher

```python
g['Publisher'].mode()
```

#### Most common platform

```python
g['Platform'].mode()
```

#### Most common genre

```python
g['Genre'].mode()
```

#### 20 highest grossing games

```python
g.nlargest(20, 'Global_Sales')[['Name', 'Global_Sales']]
```

#### Median North American sales

```python
g['NA_Sales'].median()
```

#### Approximately 10 rows within median NA_Sales

```python
x = g[g['NA_Sales'] == g['NA_Sales'].median()]
start = int(len(x) * 0.49)
end = int(len(x) * 0.51)
x[start:end]
```

#### North America best-selling number of standard deviations

```python
g.nlargest(1, 'NA_Sales')['NA_Sales'] // g['NA_Sales'].std()
```

#### Wii average global sales

```python
g[g['Platform'] == 'Wii']['Global_Sales'].mean()
```

#### All platforms average global sales

```python
for key, val in g.groupby('Platform'):
    print(key, val['Global_Sales'].mean())
```

#### Global sales by year

```python
for key, val in g.groupby('Year'):
    print(int(key), val['Global_Sales'].sum())
```

#### Global sales by Publisher and year

```python
for key, val in g.groupby(['Year', 'Publisher']):
    print(int(key[0]), key[1], val['Global_Sales'].sum())
```

```python
x = g.groupby(['Year', 'Publisher'])['Global_Sales'].sum()
df = pd.DataFrame(x)
df
```
